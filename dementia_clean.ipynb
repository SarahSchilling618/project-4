```python
conn = psycopg2.connect(**db_params)
cursor = conn.cursor()

distinct_values_query = '''
    SELECT DISTINCT "Smoking_Status"
    FROM dementia;
'''
cursor.execute(distinct_values_query)
distinct_values = cursor.fetchall()

print("Distinct values in the Smoking_Status column:")
for value in distinct_values:
    print(value[0])
```

```python
update_query = '''
    UPDATE dementia
    SET "Smoking_Status" = CASE
        WHEN "Smoking_Status" = 'Never Smoked' THEN '0'
        WHEN "Smoking_Status" = 'Former Smoker' THEN '1'
        WHEN "Smoking_Status" = 'Current Smoker' THEN '2'
        ELSE "Smoking_Status"
    END;
'''
cursor.execute(update_query)
```

```python
conn.commit()
```

```python
update_query = '''
    UPDATE dementia
    SET "APOE_4" = CASE
        WHEN "APOE_4" = 'Negative' THEN '0'
        WHEN "APOE_4" = 'Positive' THEN '1'
        ELSE "APOE_4"
    END;
'''
cursor.execute(update_query)

conn.commit()
```

```python
physical_activity_check = '''
    SELECT DISTINCT "Physical_Activity"
    FROM dementia;
'''
cursor.execute(physical_activity_check)
distinct_pa = cursor.fetchall()

for value in distinct_pa:
    print(value[0])
```

```python
update_query = '''
    UPDATE dementia
    SET "Physical_Activity" = CASE
        WHEN "Physical_Activity" = 'Sedentary' THEN '0'
        WHEN "Physical_Activity" = 'Mild Activity' THEN '1'
        WHEN "Physical_Activity" = 'Moderate Activity' THEN '2'
    END;
'''
cursor.execute(update_query)

conn.commit()
```

```python
update_query = '''
    UPDATE dementia
    SET "Depression_Status" = CASE
        WHEN "Depression_Status" = 'No' THEN '0'
        WHEN "Depression_Status" = 'Yes' THEN '1'
    END;
'''
cursor.execute(update_query)

conn.commit()
```

```python
update_query = '''
    UPDATE dementia
    SET "Medication_History" = CASE
        WHEN "Medication_History" = 'No' THEN '0'
        WHEN "Medication_History" = 'Yes' THEN '1'
    END;
'''
cursor.execute(update_query)

conn.commit()
```

```python
diet_check = '''
    SELECT DISTINCT "Nutrition_Diet"
    FROM dementia;
'''
cursor.execute(diet_check)
distinct_diet = cursor.fetchall()

for value in distinct_diet:
    print(value[0])
```

```python
update_query = '''
    UPDATE dementia
    SET "Nutrition_Diet" = CASE
        WHEN "Nutrition_Diet" = 'Low-Carb Diet' THEN '0'
        WHEN "Nutrition_Diet" = 'Mediterranean Diet' THEN '1'
        WHEN "Nutrition_Diet" = 'Balanced Diet' THEN '2'
    END;
'''
cursor.execute(update_query)

conn.commit()
```

```python
update_query = '''
    UPDATE dementia
    SET "Sleep_Quality" = CASE
        WHEN "Sleep_Quality" = 'Poor' THEN '0'
        WHEN "Sleep_Quality" = 'Good' THEN '1'
    END;
'''
cursor.execute(update_query)

conn.commit()
```

```python
health_check = '''
    SELECT DISTINCT "Chronic_Health_Conditions"
    FROM dementia;
'''
cursor.execute(health_check)
distinct_health = cursor.fetchall()

for value in distinct_health:
    print(value[0])
```

```python
update_query = '''
    UPDATE dementia
    SET "Chronic_Health_Conditions" = CASE
        WHEN "Chronic_Health_Conditions" = 'None' THEN '0'
        WHEN "Chronic_Health_Conditions" = 'Hypertension' THEN '1'
        WHEN "Chronic_Health_Conditions" = 'Diabetes' THEN '2'
        WHEN "Chronic_Health_Conditions" = 'Heart Disease' THEN '3'
    END;
'''
cursor.execute(update_query)

conn.commit()
```
