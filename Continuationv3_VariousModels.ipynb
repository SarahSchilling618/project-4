```python
import pandas as pd
from sqlalchemy import create_engine
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler, OneHotEncoder
from sklearn.metrics import confusion_matrix, accuracy_score, classification_report
from sklearn.ensemble import RandomForestClassifier
from sklearn.neighbors import KNeighborsClassifier
```

```python
# Read in the cleaned csv file from PostgreSQL_OneHotEncoder_v1.ipynb Colab Notebook

dementia_df = pd.read_csv("/content/drive/MyDrive/Bootcamp/dementia_OHC_v1.csv")
dementia_df
```

```python
dementia_df.columns
```

```python
# Reducing the number of features (columns) from the results of running Random Forest Classifier in Continuationv2_RandomForest.ipynb
dementia_df = dementia_df[['cognitive_test_scores', 'depression_status_Yes', 'apoe_4_Positive', 'dementia']]
dementia_df
```

```python

# Define the features set, by copying the dementia_df and dropping the "Dementia" column
y = dementia_df["dementia"]
X = dementia_df.copy()
X.drop("dementia", axis=1, inplace=True)
```

```python
# Preview the features data
X.head()
```

```python
# Step 3: Split the dataset into training and testing datasets by using train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1, stratify=y)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
# Creating StandardScaler instance
scaler = StandardScaler()

# Fit the training data to the standard scaler
X_scaler = scaler.fit(X_train)

# Transform the training data using the scaler
X_train_scaled = X_scaler.transform(X_train)

# Transform the testing data using the scaler
X_test_scaled = X_scaler.transform(X_test)
```

```python
# INSERT KNN
# Instantiate the model with k = 3 neighbors
model = KNeighborsClassifier(n_neighbors=3)
```

```python
# Train the model
model.fit(X_train_scaled, y_train)
```

```python
# Create predictions
knn_y_pred = model.predict(X_test_scaled)

# Review the predictions
knn_y_pred
```

```python
# Print confusion matrix. confusion_matrix(y_pred,y_test)
confusion_matrix(y_test, knn_y_pred)
```

```python
# Print classification report
target_names = ["Does Not Have Dementia [0]", "Has Dementia [1]"]
print(classification_report(y_test, knn_y_pred, target_names = target_names))
```

```python
# Support Vector Classification
Refs: Module20/Day2/Act02

# Create a support vector machine linear classifer, and fit it to the training data
from sklearn.svm import SVC
model = SVC(kernel='linear')
model.fit(X_train_scaled, y_train)
```

```python
# Module20/Day2/Act02
# Print the model score by using the test data
print("Train Acc: %.3f" % model.score(X_train_scaled, y_train))
print("Test Acc: %.3f" % model.score(X_test_scaled, y_test))
```

```python
# Module20/Day2/Act02
# Calculate the classification report

predictions = model.predict(X_test_scaled)
target_names = ["Does Not Have Dementia [0]", "Has Dementia [1]"]
results = pd.DataFrame({"Prediction": predictions, "Actual": y_test}).reset_index(drop=True)
results.head()
```

```python
print(classification_report(y_test, predictions, target_names = target_names))
```

```python
## Fit a Logistic Regression Model with the Data
# Refs: Module 20 Challenge credit-risk-classification / Module20/Day1/Act04

### Step 1: Fit a logistic regression model by using the training data (X_train_scaled and y_train).

# Import the LogisticRegression module from SKLearn
from sklearn.linear_model import LogisticRegression

# Instantiate the Logistic Regression model
# Assign a random_state parameter of 1 to the model
classifier = LogisticRegression(solver='lbfgs',
                                max_iter=500,
                                random_state=1)

# Fit the model using training data
classifier.fit(X_train_scaled, y_train)
```

```python
# Score the model
print(f"Training Data Score: {classifier.score(X_train_scaled, y_train)}")
print(f"Testing Data Score: {classifier.score(X_test_scaled, y_test)}")
```

```python
# Refs: Module 20 Challenge credit-risk-classification / Module20/Day1/Act04

## Step 2: Save the predictions on the testing data labels by using the testing feature data (`X_test_scaled`) and the fitted model.
    # Predict outcomes for test data set

lr_y_predictions = classifier.predict(X_test_scaled)
results = pd.DataFrame({"Prediction": lr_y_predictions, "Actual": y_test}).reset_index(drop=True)
results.head(10)
```

```python
# Refs: Module20/Day2/Act05
# Generating Confusion Matrix on the Logistic Regression Model
cm_lr = confusion_matrix(y_test, lr_y_predictions)
cm_df_lr = pd.DataFrame(
    cm_lr, index=["Actual 0", "Actual 1"], columns=["Predicted 0", "Predicted 1"]
)

# Calculating the accuracy score
acc_score_lr = accuracy_score(y_test, lr_y_predictions)
```

```python
# Displaying results
print("Confusion Matrix")
display(cm_df_lr)
print(f"Accuracy Score : {acc_score_lr}")
print("Classification Report")
print(classification_report(y_test, lr_y_predictions))
```
